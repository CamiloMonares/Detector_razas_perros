```python
# Función que retorna True si un perro es detectado en la imagen
def detector_perros(ruta_imagen):
    indice = VGG16_predict(ruta_imagen)
    
    if indice >= 151 and indice <= 268:
        return True
    
    return False
```

```python
import torch
import os
from torchvision import datasets, transforms

def carga_datos(data_dir, batch_size, training_mean, training_std_deviation, es_set_entrenamiento = False):
    transformada = []
    
    if es_set_entrenamiento:
        transformada = [
            transforms.Resize((224,224)),
        ]
    else:
        transformada = [
            transforms.RandomRotation(30),
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
        ]
    transformada.append(transforms.ToTensor())
    transformada.append(transforms.Normalize(training_mean, training_std_deviation))
    
    transformada = transforms.Compose(transformada)
        
    dataset = datasets.ImageFolder(data_dir, transform=transformada)
    return torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
batch_size = 20
train_mean = [0.485, 0.456, 0.406]
train_std = [0.229, 0.224, 0.225]
carga_entrenamiento = carga_datos('C:\\Desarrollo Trabajo de titulo 2\\parapy\\Data\\dogimages\\train', batch_size, train_mean, train_std)
carga_validacion = carga_datos('C:\\Desarrollo Trabajo de titulo 2\\parapy\\Data\\dogimages\\valid', batch_size, train_mean, train_std)
carga_test = carga_datos('C:\\Desarrollo Trabajo de titulo 2\\parapy\\Data\\dogimages\\test', batch_size, train_mean, train_std, True)

cargar_transfer = {
    'entrenamiento': carga_entrenamiento,
    'validacion': carga_validacion,
    'test': carga_test
}
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

usar_cuda = torch.cuda.is_available()

def cargar_modelo_transfer(modelo_guardado = None):
    modelo_transfer = models.vgg16(pretrained=True)

    modelo_transfer.classifier[6] = nn.Linear(modelo_transfer.classifier[6].in_features, 133)

    # freeze the parameters
    for parametro in modelo_transfer.features.parameters():
        parametro.requires_grad = False
    
    if modelo_guardado != None:
        modelo_transfer.load_state_dict(torch.load(modelo_guardado))
        
    if usar_cuda:
        modelo_transfer = modelo_transfer.cuda()
    
    return modelo_transfer
```

```python
modelo_transfer = cargar_modelo_transfer()
```

```python
import torch.optim as optim
criterio = nn.CrossEntropyLoss()
optimizador_transfer = optim.SGD(modelo_transfer.classifier.parameters(), lr=0.001)
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

def entrenar(n_epochs, loaders, model, optimizer, criterion, use_cuda, save_path):
    # inicializamos la variable para el minimo error
    min_error_validacion = np.Inf 
    
    for epoch in range(1, n_epochs+1):
        # inicializamos las variables para monitorear el error de entrenamiento y validación

        error_entrenamiento = 0.0
        error_validacion = 0.0
        
        # entrenamiento del modelo
 
        model.train()
        for batch_idx, (data, target) in enumerate(loaders['entrenamiento']):
            # usamos GPU
            if usar_cuda:
                data, target = data.cuda(), target.cuda()
            
            optimizer.zero_grad()
            
            output = model(data)
            
            # se calcula la perdida de lotes
            perdida = criterion(output, target)
            
            #backward pass: se calcula la perdida de gradiente con respecto a los parámetros del modelo
            perdida.backward()
            
            # se actualizan los parametros
            optimizer.step()
            error_entrenamiento = error_entrenamiento + ((1 / (batch_idx + 1)) * (perdida.data - error_entrenamiento))
            
            
        # validacion del modelo
        model.eval()
        for batch_idx, (data, target) in enumerate(loaders['validacion']):
            # usamos GPU
            if usar_cuda:
                data, target = data.cuda(), target.cuda()
            
            output = model(data)
            
            perdida = criterion(output, target)
            
            error_validacion = error_validacion + ((1 / (batch_idx + 1)) * (perdida.data - error_validacion))
            
        print('Epoch: {} \tError entrenamiento: {:.6f} \tError validacion: {:.6f}'.format(
            epoch, 
            error_entrenamiento,
            error_validacion
            ))
        
        # guardamos el modelo si el error de validación ha disminuido
        if error_validacion <= min_error_validacion:
            print('ha disminuido el error de validación desde : {:.6f} --> {:.6f}. Guardando modelo...'.format(
            min_error_validacion, 
            error_validacion))
            torch.save(model.state_dict(), save_path)
            min_error_validacion = error_validacion
            
    # Retorna el modelo entrenado
    return model
```

```python
# Entrenamiento del modelo
modelo_transfer = entrenar(10, cargar_transfer, modelo_transfer, optimizador_transfer, criterio, usar_cuda, 'model_transfer.pt')

# Carga del modelo que obtuvo la mejor precisión en la validación
modelo_transfer.load_state_dict(torch.load('model_transfer.pt'))
```

```python
def test(loaders, model, criterion, usar_cuda):

    # Monitoreamos el error y la precisión del modelo
    error_test = 0.
    correcto = 0.
    total = 0.

    model.eval()
    for batch_idx, (data, target) in enumerate(loaders['test']):
        # Usamos GPU
        if usar_cuda:
            data, target = data.cuda(), target.cuda()
        output = model(data)
        # Calculo del error
        error = criterion(output, target)
        # Actualizamos el error promedio 
        error_test = error_test + ((1 / (batch_idx + 1)) * (error.data - error_test))
        # Se convierten las probabilidades de salida en la clase predicha
        pred = output.data.max(1, keepdim=True)[1]
        # Comparación entre la predicción con las etiquetas reales
        correcto += np.sum(np.squeeze(pred.eq(target.data.view_as(pred))).cpu().numpy())
        total += data.size(0)
            
    print('Error en el test: {:.6f}\n'.format(error_test))

    print('\nPrecisión en el test: %2d%% (%2d/%2d)' % (100. * correcto / total, correcto, total))
```

```python
test(cargar_transfer, modelo_transfer, criterio, usar_cuda)
```
